```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
df=pd.read_csv('Amazon Scraping.csv')
```

```python
df
```

```python
x=df.Asin.to_list()
y=df.country.to_list()
```

```python
urls=[]
for i,j in zip(x,y):
    if i.endswith('X'):
        print(f"https://www.amazon.{j}/dp/{i}")
        urls.append((f"https://www.amazon.{j}/dp/{i}"))
    
```

```python
header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36" ,
    'referer':'https://www.google.com/'
}

lists=[]
for u in range(len(urls)):

    a = requests.get(urls[u],headers=header)

    soup=BeautifulSoup(a.content,'html.parser') #storing the content of HTML in a variable


    
    spacing=['a-spacing-none a-text-normal','a-size-extra-large','a-size-large product-title-word-break']
    price=['a-price-whole','a-size-base a-color-price a-color-price','a-button a-button-selected a-spacing-mini a-button-toggle format','a-color-base','a-button a-button-selected a-spacing-mini a-button-toggle format']
    details=['a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list','a-expander-content a-expander-section-content a-section-expander-inner',]
    
    
    for i in spacing:
        
        for d in range(u):
            info={}
            info_box=soup.find(class_=i)
            if info_box !=None:
                info['title']=info_box.text


        for j in soup.find_all('img'):
            s=j.get('src')
            if s.endswith('ML2_.jpg') or s.endswith('200_.jpg'):
                info['Image url']=s

        for k in price:
            info_box=soup.find(class_=k)
            if info_box !=None:
                test=info_box.text
                info['Price of the Product']=test        


        for l in details:
            info_box=soup.find(class_=l)
            if info_box !=None:
                d=info_box.text
                d=d.replace('\n','')
                d=d.encode("ascii", "ignore")
                info['Product Details']=d
        info.update({'info':d})
    lists.append(info)

```
